## 什么是多项式回归

在机器学习算法上, 还是采用了线性回归的思路
它的关键在于, 我们为原来的样本添加的新的特征, 而我们得到新特征的方式是原有特征的线性组合, 来解决一些非线性问题

这个思路是升维, 添加新特征, 使得我们的算法可以更好的拟合高维度的数据, 后面学习SVM的时候会再次强调
PCA是降维

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)  # 转换成2维数组 
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, 100)
```

```python
plt.scatter(x, y)
plt.show()
```

### 线性回归？

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_predict = lin_reg.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(x, y_predict, color='r')
plt.show()
```

### 解决方案， 添加一个特征

```python
X2 = np.hstack([X, X**2])
```

```python
X2.shape
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X2, y)
y_predict2 = lin_reg2.predict(X2)
```

```python
# 错误示范: 原因是x的顺序是无序的, 我们需要给它按顺序排列一下
plt.scatter(x, y)
plt.plot(x, y_predict2, color='r')
plt.show()
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y_predict2[np.argsort(x)], color='r')  # 传进的是排序过的x
plt.show()
```

```python
lin_reg2.coef_  # 第一个是x的系数, 第二个是x**2的系数
```

```python
lin_reg2.intercept_  # 截距
```
