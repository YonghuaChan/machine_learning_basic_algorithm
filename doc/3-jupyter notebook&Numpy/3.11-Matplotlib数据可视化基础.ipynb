## <font size=3>matplotlib 基础</font>

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np
```

### <font size=3>折线图</font>

```python
'''向量'''
x = np.linspace(0, 10, 100)
x
```

```python
siny = np.sin(x)
siny
```

```python
plt.plot(x, siny)
```

```python
plt.show()
```

```python
cosy = np.cos(x)
cosy
```

```python
plt.plot(x, cosy)
```

```python
'''
两个曲线

修改颜色

'''
plt.plot(x, siny, color='#B54434')
plt.plot(x, cosy, color='green')
plt.show()
```

```python
'''
样式修改 linestyle
'''

plt.plot(x, siny, color='#B54434', linestyle=':')
plt.plot(x, cosy, color='green', linestyle = '--')
plt.show()
```

```python
'''
添加坐标轴
'''

plt.plot(x, siny, color='#B54434', linestyle=':')
plt.plot(x, cosy, color='green', linestyle = '--')
plt.xlim(-5, 15)
plt.show()
```

```python
'''
添加坐标轴 xlim, ylim
'''

plt.plot(x, siny, color='#B54434', linestyle=':')
plt.plot(x, cosy, color='green', linestyle = '--')
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.show()
```

```python
'''
添加坐标轴 plt.axis
'''

plt.plot(x, siny, color='#B54434', linestyle=':')
plt.plot(x, cosy, color='green', linestyle = '--')
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
'''
添加坐标轴标签 plt.xlabel/ylabel
'''

plt.plot(x, siny, color='#B54434', linestyle=':')
plt.plot(x, cosy, color='green', linestyle = '--')
plt.axis([-1, 11, -2, 2])
plt.xlabel('x axis')
plt.ylabel('y value')
plt.show()
```

```python
'''
添加图像图示
例如:
1. 在plt.plot(label="sin(x)")
2. plt.legend(): 加上标签
'''

plt.plot(x, siny, color='#B54434', label="sin(x)")  # label: 曲线标签名
plt.plot(x, cosy, color='green', linestyle = '--', label="cos(x)")
plt.axis([-1, 11, -2, 2])
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()  # 加上标签
plt.show()
```

```python
'''
标题
'''

plt.plot(x, siny, color='#B54434', label="sin(x)")  # label: 曲线标签名
plt.plot(x, cosy, color='green', linestyle = '--', label="cos(x)")
plt.axis([-1, 11, -2, 2])
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()  # 加上标签
plt.title('Welcome to mechine learning world')
plt.show()
```

### <font size=3>Scatter Plot: 散点图</font>

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color='#B54434')
plt.show()
```

```python
'''
对于折线图来说, 一般横轴是特征, 而纵轴是取值

但对于散点图来说并非如此, 横纵两个周都是特征
    一般我们在二维特征的时候使用散点图
    把点打印在图像上, 不同的颜色表示不同的label

'''
```

```python
'''
二维的正态分布

plt.scatter(x, y)
    - alpha: 不透明度, 0代表完全透明

'''
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

plt.scatter(x, y, alpha=0.4)
plt.show()
```
